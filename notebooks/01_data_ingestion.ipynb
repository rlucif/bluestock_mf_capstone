```python
from pathlib import Path
import pandas as pd

raw_path = Path("../data/raw")
csv_files = sorted(raw_path.glob("*.csv"))
datasets = {}

for file in csv_files:
    df = pd.read_csv(file)
    datasets[file.stem] = df

    print("=" * 80)
    print(file.name)
    print("=" * 80)

    print("\nSHAPE")
    print(df.shape)

    print("\nDTYPES")
    print(df.dtypes)

    print("\nHEAD")
    print(df.head())

    print("\n\n")
```

```python
print(f"Total datasets loaded: {len(datasets)}")
print("\nDatasets:")
for name in datasets.keys():
    print(name)
```
